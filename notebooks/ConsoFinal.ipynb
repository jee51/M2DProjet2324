# Consommation des moteurs.

```python
import os
from glob import glob
import numpy as np
import pandas as pd
import statsmodels.formula.api as smf
import tabata as tbt
from tabata import Opset
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

%reload_ext autoreload
%autoreload 2
```

```python
datadir = "../data/"
datafiles = glob(os.path.join(datadir,'*.h5'))
sorting = lambda name: int(name[8:])
```

```python
phaselist = ['climb', 'cruise', 'descend']
```

```python
from utils import get_consumption

R = []
df = pd.DataFrame()
for acfile in datafiles:
    ac = Opset(acfile, sortkey=sorting)
    for phase in phaselist:
        df = get_consumption(ac, phase=phase)
        df['phase'] = phase
        df['AC'] = 'AC'+ac.storename[-5:-3]
        cols = df.columns
        df = df[['AC','ENG','Flight','phase'] + list(cols[3:-1])]
        R += [df]
        break
    break
df = pd.concat(R)
```
